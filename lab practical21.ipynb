```python
import pandas as pd
sales = pd.read_csv(r"c:\Users\LENOVO\Downloads\sales_data.csv") 
print(sales)
```

```python
# 2. Create Total_Revenue column

sales["Total_Revenue"] = sales["Quantity"] * sales["Price"]

print(sales.head())
```

```python
# 3. Total revenue generated for each product
print("Total revenue for each product:")
print(sales.groupby("Product")["Total_Revenue"].sum())
```

```python
# 4. Total units sold for each category
print("\nTotal units sold for each category:")
print(sales.groupby("Category")["Quantity"].sum())
```

```python
# 5. Average price per unit for each category
print("\nAverage price per unit for each category:")
print(sales.groupby("Category")["Price"].mean())
```

```python
# 6. Product with maximum units sold
print("\nProduct with maximum units sold:")
print(sales.loc[sales["Quantity"].idxmax()])
```

```python
# 7. Total revenue for each month
print("\nTotal revenue for each month:")
print(sales.groupby("Month")["Total_Revenue"].sum())
```

```python
# 8. Products where units sold > 40
print("\nProducts with Quantity greater than 40:")
print(sales[sales["Quantity"] > 40])
```

```python
# 9. Sort dataset by Total_Revenue in descending order
print("\nDataset sorted by Total_Revenue:")
print(sales.sort_values(by="Total_Revenue", ascending=False))
```
