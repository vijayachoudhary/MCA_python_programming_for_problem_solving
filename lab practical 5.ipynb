```python
text = input(" enter sentence:")
print("Entered sentences:",text)
```

```python
def reverse_vowels(s):
    vowels = "aeiouAEIOU"
    s_list = list(s)
    vowel_chars = [ch for ch in s_list if ch in vowels]
    vowel_chars.reverse()
    j = 0
    for i in range(len(s_list)):
        if s_list[i] in vowels:
            s_list[i] = vowel_chars[j]
            j += 1
    return "".join(s_list)
print("Output:", reverse_vowels(text))
```
